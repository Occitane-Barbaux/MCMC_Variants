```python
import numpy as np
import re 
import os 
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
plt.rcParams['figure.figsize'] = [7, 7]
import seaborn as sns
sns.set_context("talk")
```

```python
pathInp  = "/cnrm/amacs/USERS/barbauxo/SQR_Full"
#pathInp="/home/barbauxo/Documents/Doctorat/03_Travail/2023_01 Application Tricastin/Data/Observations"
ListeNC=os.listdir(pathInp)

r = re.compile(".*nc")
ListeNC = list(filter(r.match, ListeNC)) 

id=[]
maxTX=[]
maxDate=[]
maxprof=[]
#POur chaque station extrait le max de la serie et la date correspondante"
for NC in ListeNC:

  StationNC = xr.open_dataset(os.path.join(pathInp,NC))
  Station = pd.DataFrame( {"TX":StationNC.TX.values.squeeze(),"Time":StationNC.time["time"].values}  )
    
  if sum(np.isfinite(Station["TX"]))!=0 :
    id.append(np.int(re.findall('[0-9]+', NC)[0]))
    maxStation=np.nanargmax(Station["TX"])
    maxTX.append(Station["TX"][maxStation])
    maxDate.append(Station["Time"][maxStation])
    if Station["TX"][maxStation]>50:
      print(NC)
    Station=Station[(np.isfinite(Station.TX))]
    years=Station.Time.values.astype('datetime64[Y]').astype(int)+1970
    
    prof=max(years)-min(years)
    maxprof.append(prof)
Stations_Values_TX= pd.DataFrame( {"NUM_POSTE":id, "TX":maxTX,"Date":maxDate, "Profondeur":maxprof}  )
```

```python
#Ajout coordonnées
Stations_List_TX=pd.read_excel("/home/barbauxo/Documents/Doctorat/03_Travail/Préparation des données/Observations/Météo France/Stations d'Observations/Scripts extractions/SeriesHomogeneiseesBDCLIM_TT_lsdh_SQR_20211110_v2.ods",'tx_metropole')
Stations_List_TX=Stations_List_TX[['NUM_POSTE',"COMMUNE",'lat(dms)','lon(dms)','lambx(hm)', 'lamby(hm)','altitude(m)',]]


Stations_TX=pd.merge(Stations_Values_TX,Stations_List_TX, on=["NUM_POSTE"])
```

```python

def Coordinate_dms_to_fraction(Station_TX, origin, new):
  Stations_TX[new]=[0] * len(Stations_TX)
  for i in range (0, len(Stations_TX)):
    neg=False
    coo=Stations_TX[origin][i]
    if (coo<0):
      neg=True
    coo=abs(coo)
    coordinate= (coo%100)/360
    coo=(coo-coo%100)/100
    coordinate= coordinate + (coo%100)/60
    coo=(coo-coo%100)/100
    coordinate= coordinate + coo
    if neg:
      coordinate=-coordinate
    Stations_TX[new][i]= coordinate
  return Stations_TX


Stations_TX=Coordinate_dms_to_fraction(Stations_TX, 'lon(dms)', 'longitude')
Stations_TX=Coordinate_dms_to_fraction(Stations_TX, 'lat(dms)', 'latitude')

```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent([-6, 10, 41, 52])
ax.add_feature(cfeature.COASTLINE.with_scale('50m'))

ax.add_feature(cfeature.BORDERS.with_scale('50m'), linestyle='-')

ax.get_legend()
norm = mcolors.TwoSlopeNorm(vmin=25,vmax=45,vcenter=40)
#norm = mcolors.BoundaryNorm([25, 30, 35, 40,45],4)

plt.scatter( Stations_TX["longitude"],Stations_TX["latitude"], c=Stations_TX["TX"], ec='k',cmap='hot_r',norm=norm)
cbar=plt.colorbar( orientation="horizontal", pad=0.05,ticks=[25,30,35,40,45])



ax.set(title='Records de température Maximale \n quotidienne mesurées en France')
plt.show()
```

```python
## Tricastin
## 44° 19′ 51″ N, 4° 43′ 53″ E
# Altitude 54m sur GG earth

Lat_Tri=44+19/60+51/(60*60)
Lon_Tri=4+43/60+53/(60*60)
Alt_Tri=54
Lon_Tri,Lat_Tri
```

```python
def Distance(Stations_TX, Lat_Tri, Lon_Tri):
  Stations_TX["distance (km)"]=[0] * len(Stations_TX)
  for i in range (0, len(Stations_TX)):
    dist=6378.137*np.arccos(np.sin(Lat_Tri*np.pi/180)*np.sin(Stations_TX["latitude"][i]*np.pi/180)+np.cos(Lat_Tri*np.pi/180)*np.cos(Stations_TX["latitude"][i]*np.pi/180)*np.cos((Lon_Tri-Stations_TX["longitude"][i])*np.pi/180))
    Stations_TX["distance (km)"][i]= dist
  return Stations_TX

Stations_TX=Distance(Stations_TX, Lat_Tri, Lon_Tri)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent([-6, 10, 41, 52])
ax.add_feature(cfeature.COASTLINE.with_scale('50m'))

ax.add_feature(cfeature.BORDERS.with_scale('50m'), linestyle='-')

ax.get_legend()
norm = mcolors.TwoSlopeNorm(vmin=10,vmax=100,vcenter=50)
#norm = mcolors.BoundaryNorm([25, 30, 35, 40,45],4)

plt.scatter( Stations_TX["longitude"],Stations_TX["latitude"], c=Stations_TX["distance (km)"], ec='k',cmap='hot_r',norm=norm)
cbar=plt.colorbar( orientation="horizontal", pad=0.05,ticks=[10,50,100])

plt.scatter(Lon_Tri, Lat_Tri, c='lime', marker="s",edgecolor='black', s=30)

ax.set(title='Distance au site de Tricastin (km)')
plt.show()
```

### Automatisation

```python
n_station=str(choix["NUM_POSTE"].iloc[0])
nom=str(choix["COMMUNE"].iloc[0]).title()

def present_station(n_station,nom,Ruptures):
    r = re.compile(n_station)
    nc_num=list(filter(r.match, ListeNC))
    StationNC = xr.open_dataset(os.path.join(pathInp,nc_num[0]))
    Station = pd.DataFrame( {"TX":StationNC.TX.values.squeeze(),"Time":StationNC.time["time"].values}  )
    print(Station.head())
    
    fig, ax = plt.subplots()
    ax.plot( Station.Time , Station.TX , color = "red" , linestyle = "-" , marker = "" )
    
    ax.vlines(Ruptures.Time[0],0, 1,transform=ax.get_xaxis_transform(), colors='black',label="Ruptures")
    for i in range(1,len(Ruptures)):
        ax.vlines(Ruptures.Time[i],0, 1,transform=ax.get_xaxis_transform(), colors='black')

    plt.title(r"Température mesurée à la station de " + nom)
    plt.legend()
    plt.show()
    plt.clf()
    years=(Station.Time.values.astype('datetime64[Y]').astype(int)+1970)
    year=[]
    maxTX=[]
    for i in np.unique(years):
        year.append(i)
        if (len(Station[(Station.Time.values.astype('datetime64[Y]').astype(int)+1970)==i].TX)>150):
            maxTX.append(max(Station[(Station.Time.values.astype('datetime64[Y]').astype(int)+1970)==i].TX))
        else:
            maxTX.append(np.float("nan"))
            
    Max_Annuels=pd.DataFrame( {"Year":year, "TX":maxTX}  )
    
    idx = np.isfinite(Max_Annuels.TX)
    z = np.polyfit(Max_Annuels.Year[idx], Max_Annuels.TX[idx], 1)
    p = np.poly1d(z)
    
    fig, ax = plt.subplots()
    plt.ylim([np.min(Max_Annuels.TX)-1, np.max(Max_Annuels.TX)+1])
    ax.plot( Max_Annuels.Year , Max_Annuels.TX , color = "red" , linestyle = "-" , marker = "s" )
    ax.plot(Max_Annuels.Year[idx],p(Max_Annuels.Year[idx]),"--",label=r"Tendance " +str(np.round(z[0]*10,2))+"°C par décade")

    
    plt.title(r"Maxima Annuels à la station de "+nom)
    ax.vlines(Ruptures.Year[0],0, 1,transform=ax.get_xaxis_transform(), colors='black',label="Ruptures")
    for i in range(1,len(Ruptures)):
        ax.vlines(Ruptures.Year[i],0, 1,transform=ax.get_xaxis_transform(), colors='black')
    plt.legend()
    plt.show()
    return Max_Annuels
Station_Orange=present_station(n_station,nom,Ruptures)
```
